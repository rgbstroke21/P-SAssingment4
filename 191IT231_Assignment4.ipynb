```python
# Import Necessary Libraries
import random
import numpy as np
import pandas as pd
import random as rd
```

```python
# form words from the input text file and we are gonna use these words as our states
def formWords( data ):
    data = data.lower()
#     data = data.replace(',','')
#     data = data.replace('.','')
    words = data.split()
    return words
```

```python
uniqueWords = list( set(words) )
```

```python
# Store the words in a matrix and create the transition matrix
matrix = pd.DataFrame( data = 0 , index = uniqueWords , columns=uniqueWords , dtype = int)
```

```python
# returns [ maximum freq which a phrase occur , the phrase , all possible phrases]
def completeMySentence( storeLastWord , uniqueWords , matrix , order ):
    
    if not storeLastWord in uniqueWords:
        return [0,0,0]
    
    remainingSentences = {}
    
    for j in range(1000):
        
        curSentence = ''
        lastWord = storeLastWord
        for i in range(order):
            if not lastWord in uniqueWords:
                sentenceCompleted = False
                break
            nextWord = random.choices( uniqueWords , weights = matrix[lastWord].values , k = 1 )[0]
            curSentence += ' ' + nextWord
            lastWord = nextWord

        if not ( curSentence in remainingSentences.keys() ):
            remainingSentences[curSentence] = 0
        remainingSentences[curSentence] += 1

    #     if not sentenceCompleted:
    #         print("Can't complete the sentence partial results are shown below")
    
    
    
#     print('Following appear in 1000 searchs : ')
#     for entry in remainingSentences:
#         print( 'Frequency = ' , remainingSentences[entry] ,'\t', sentence + entry )


    maxFreq = max( remainingSentences.values() )
    print("\n\nMost probable result occured ", maxFreq , "times" )
    
    mostProbableCompletion = ''
    for entry in remainingSentences:
        if remainingSentences[entry] == maxFreq:
            mostProbableCompletion = entry
            break
    
    answer = [maxFreq , mostProbableCompletion , remainingSentences]
    
    
    return answer
```

```python
def getLastWord( string ):
    wordsInString = string.split()
    if len(wordsInString) == 0:
        return ''
    lastWord = wordsInString[-1].lower()
    return lastWord
```

```python
while True:
    sentence = input("\nEnter the sentence\n")
    order = int( input("Enter how many words to add\n") )
    storeLastWord = getLastWord(sentence)

    print('\nLast Word is ' , storeLastWord )

    results = completeMySentence( storeLastWord , uniqueWords , matrix , order )

    if results[0]==0:
        print("No matching words found\n")
        continue
    
    print('\nThe best result occured ' , results[0]  , ' among 1000 attempts : ' )
    print( sentence+results[1])
    
    seeAll = 0
    seeAll = int( input('\n1 . See all the results \n2 . Continue \n' ) )
    if seeAll == 1:
        print("\n\nPrinting all the results : ")
        for entry in results[2]:
            print( 'Freq = ' , results[2][entry] , '\t\t\t' , entry )
    
    stopOrContinue = int( input("1 : Enter another sentence.\n2 : Stop\n") )
    if stopOrContinue!=1:
        break
```
